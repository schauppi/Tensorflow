```python
import yfinance as yf
import matplotlib.pyplot as plt
import mplfinance as mpf
import pandas as pd
```

```python
bitcoin = yf.Ticker("BTC-EUR")
```

```python
#bitcoin.info
```

```python
#hist_btc = bitcoin.history(start="2014-09-17")
hist_btc = bitcoin.history(period="max")
```

```python
hist_btc.head()
```

```python
btc_data = hist_btc[["Open", "High", "Low", "Close", "Volume"]]
```

```python
btc_data.head()
btc_data.info()
#btc_data = btc_data.head()
```

```python
mpf.plot(btc_data, type="candle", style="charles", title="Bitcoin", ylabel="Price ", volume=True, ylabel_lower="Market Volume", mav=2, figratio=(24,10))
```
