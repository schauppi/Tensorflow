```python
import yfinance as yf
import matplotlib.pyplot as plt
import mplfinance as mpf
import pandas as pd
from datetime import date
import numpy as np
```

```python
#Dict for Stock names and abbreviations
stocks_dict = {"btc":["BTC-EUR", "Bitcoin"], "atx":["^ATX", "Austrian Traded Index"], "bldp":["BLDP", "Ballard Power Systems"], 
               "nel":["D7G.F", "Nel Asa"], "plug":["PLUG", "Plug Power"], "nvda":["NVDA", "Nvidia"], "voe":["VOE.VI", "Voestalpine"], 
               "tesla":["TL0.F", "Tesla Motors"], "intel":["INTC", "Intel"], "apple":["AAPL", "Apple"], "alib":["AHLA.F", "Alibiba"],
               "alphab":["ABEA.F", "Alphabet"]}

#lists for stock data
stocks_fetched = []
history_fetched = []
stocks_data = []

#actual date
date = date.today()
```

```python
#Function for fetching Stock Data Infos
def fetch_stock_data(abbreviation):
    fetched = yf.Ticker(abbreviation)
    return fetched

#Function for fetching the historical Data of the Stock
def get_history(name):
    #hist = name.history(start=2018-10-10, end=date)
    #hist = name.history(period="5y")
    hist = name.history(period="max")
    return hist

#Function for fetching values 
def get_values(hist):
    vals = hist[["Open", "High", "Low", "Close", "Volume"]]
    return vals
```

```python
#Iterate to get stock data
i = 0
for key in stocks_dict:
    stocks_fetched.append(fetch_stock_data(stocks_dict[key][0]))
    history_fetched.append(get_history(stocks_fetched[i]))
    stocks_data.append(get_values(history_fetched[i]))
    i += 1
```

```python
#scale the data
#bring it in shape (samples, features)
from sklearn.preprocessing import RobustScaler
scaler = RobustScaler()
scaler = scaler.fit(bitcoin.to_numpy())
train_data = scaler.transform(bitcoin.to_numpy())
```

```python
import tensorflow as tf
from tensorflow import keras

def windowed_dataset(series, window_size, batch_size, shuffle_buffer):
    series = tf.expand_dims(series, axis=-1)
    dataset = tf.data.Dataset.from_tensor_slices(series)
    dataset = dataset.window(window_size + 1, shift = 1, drop_remainder = True)
    dataset = dataset.flat_map(lambda window: window.batch(window_size + 1))
    dataset = dataset.map(lambda window: (window[:-1], window[-1]))
    dataset = dataset.shuffle(shuffle_buffer)
    dataset = dataset.batch(batch_size).prefetch(1)
    return dataset
```

```python
window_size = 30
batch_size = 32
shuffle_buffer_size = 1000
train_set = windowed_dataset(train_data, window_size, batch_size, shuffle_buffer_size)
```
