```python
import yfinance as yf
import matplotlib.pyplot as plt
import mplfinance as mpf
import pandas as pd
```

```python
#Dict for Stock names and abbreviations
stocks_dict = {"btc":"BTC-EUR", "atx":"^ATX", "bldp":"BLDP", "nel":"D7G.F", 
          "plug":"PLUG", "nvda":"NVDA", "voe":"VOE.VI", "tesla":"TL0.F", 
          "intel":"INTC", "apple":"AAPL", "alib":"AHLA.F", "alphab":"ABEA.F"}

stocks_names =["Bitcoin", "Austrian Traded Index", "Ballard Power Systems", "Nel ASA",
              "Plug Power", "Nvidia", "Voestalpine", "Tesla", "Intel", "Apple", "Alibaba", "Alphabet"] 

#lists for stock data
stocks_fetched = []
history_fetched = []
stocks_data = []

#ToDo: names to dictionaries 
```

```python
#Function for fetching Stock Data Infos
def fetch_stock_data(abbreviation):
    fetched = yf.Ticker(abbreviation)
    return fetched

#Function for fetching the historical Data of the Stock
def get_history(name):
    hist = name.history(period="max")
    return hist

#Function for fetching values 
def get_values(hist):
    vals = hist[["Open", "High", "Low", "Close", "Volume"]]
    return vals
```

```python
#Iterate to get stock data
i = 0
for key in stocks_dict:
    stocks_fetched.append(fetch_stock_data(stocks_dict[key]))
    history_fetched.append(get_history(stocks_fetched[i]))
    stocks_data.append(get_values(history_fetched[i]))
    i += 1
```

```python
mpf.plot(stocks_data[0], type="candle", style="charles", title="Ausrian Traded Index", ylabel="Price ", volume=True, ylabel_lower="Market Volume", mav=2, figratio=(24,10))
```
